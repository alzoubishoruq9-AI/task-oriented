```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.tools import StructuredTool ,tool

from pydantic import BaseModel , Field
from langchain.agents import create_agent
```

```python
load_dotenv()
os.environ["OPEN_API_KEY"]=os.getenv("OPEN_API_KEY")
llm=ChatOpenAI(model="gpt-4o-mini",max_tokens=1000,temperature=0)
```

```python
def summarieze(text):
  prompt=PromptTemplate(
      input_variables=["text"],
      template="Summarize the following text:\n\n{text}\n\nSummary:"
  )
  chain=prompt|llm
  return chain.invoke({"text":text}).content
def translate(text):
  prompt=PromptTemplate(
      input_variables=["text"],
      template="Translate the following text to Spanish:\n\n{text}\n\nTranslation:"
  )
  chain=prompt|llm
  return chain.invoke({"text":text}).content

```

```python
class TextInput(BaseModel):
  text:str=Field(description="The text to summarize or translate")
```

```python
tools=[
    StructuredTool.from_function(
        name="summarize",
        func=summarieze,
        description="Useful for summarizing text",
        args_schema=TextInput
    ),
    StructuredTool.from_function(
        name="translate",
        func=translate,
        description="Useful for translating text to Spanish",
        args_schema=TextInput
    )


]
```

```python
prompt = PromptTemplate(
    input_variables=["input"],
    template="""
Summarize the following text and then translate the summary to Spanish:

Text: {input}

Use the following steps:
1. Use the Summarize tool to summarize the text.
2. Use the Translate tool to translate the summary to Spanish.
3. Provide the final result in this format:
   Summary (English): [English summary]
   Translation (Spanish): [Spanish translation]
"""
)
```

```python
agent = create_agent(
    model=llm,
    tools=tools,
    system_prompt=prompt.template
)
```

```python
def run_agent_query(agent,query):
  result=create_agent.invoke({"input":query})
  return result["output"]
```

```python
query = """The quick brown fox jumps over the lazy dog. This sentence is often used as a pangram in typography
to display font examples, as it contains every letter of the English alphabet. However, it's not the only pangram
in existence. Another example is 'Pack my box with five dozen liquor jugs', which is shorter but less commonly used."""
result=run_agent_query(agent,query)
print("\nQuery:")
print(query)
print("\nResult:")
print(result)
```
